```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
url = 'https://www.etnews.com/news/section.html?id1=04&id2=044&page=56'
html = urlopen(url)
soup = bs(html.read(),"html.parser")
last_page = soup.find('a',{'href':'javascript:;'}).text
crawl_df = pd.DataFrame(columns=["title",'context'])
for page in range(1,int(last_page)+1):
  url = "https://www.etnews.com/news/section.html?id1=04&id2=044&page="+str(page)
  html = urlopen(url)
  soup = bs(html.read(),"html.parser")
  base_url = "https://www.etnews.com"
  print(f'{page}진행중')

  link_url = []
  for links in soup.select('p > a'):
    link = links.attrs['href']
    if "https" not in link:
      link_url.append(base_url+link)
  for link in link_url:
    html = urlopen(link)
    soup = bs(html.read(),'html.parser')
    title = soup.find_all('h2')[0].text
    context = soup.findAll('p')[0].text.replace("\n","")
    tmp_df = pd.DataFrame([[title,context]],columns=crawl_df.columns)
    crawl_df = pd.concat([crawl_df, tmp_df])


```

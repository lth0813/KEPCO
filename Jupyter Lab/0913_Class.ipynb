```python
import pandas as pd
import numpy as np
```

```python
# 평균 = 모든 데이터를 합해서 데이터의 개수로 나눔
# 파이썬 자체로 평균 구하기
X = [1,2,3,4]
s = 0
for x in X:
    s += x
s/len(X)
```

```python
# numpy 이용해서 평균 구하기
X = np.array(X) # numpy는 array 형식으로 바꿔주는게 좋다.
np.sum(X)/len(X)
# np.mean(X)
```

```python
X = [1,2,3,4]
np.sum(X)/len(X)
np.mean(X)
# X.mean() - 리스트 자체에는 .mean()이라는 함수가 없어서 np.array로 바꿔줘야함함 
np.array(X).mean()
```

```python
# pandas 이용해서 평균 구하기
pd.Series(X).sum()/len(pd.Series(X)) # 시리즈 메서드에 있는 .sum 과 .mean을 사용
pd.Series(X).mean()
```

```python
# numpy 중앙값 구하기
np.median(X) # numpy
# np.array(X).median()
```

```python
dir(np.array) # 함수
dir(np.array([1])) # ndarray
dir(np.ndarray) # ndarray
set(dir(np)) - set(dir(np.ndarray)) # ndarray에는 median 메서드가 없는거 확인
```

```python
# pandas 중앙값 구하기
pd.Series(X).median()
```

```python
X = pd.Series(np.random.randint(10,100,1000))
```

```python
# 중복값 제외하고 중앙값 구하기
X.duplicated() # 중복값 지우기 = bool형태
X[~X.duplicated()].median() # True값만 가지고 중앙값 출력
X.drop_duplicates().median() # 중복값 제거하고 중앙값 출력
```

```python
# numpy / pandas 최솟값, 최댓값 구하기
np.max(X)
X.max()
np.min(X)
X.min()
```

```python
# numpy / pandas 분위수 구하기
np.quantile(X,0.5) # = np.percentile(X,50)
X.quantile(0.5)
```

```python
# numpy / pandas 분산(VAR) 구하기 (루트(VAR)는 표준편차), 분산 = 떨어진 정도
np.sqrt(np.sum(np.square((X - X.mean())))/X.size-1) # 메서드 안쓰고 수식대로 구하기
X_var = np.var(X) # 분산
np.sqrt(X_var) # 표준편차 (sqrt = 루트 씌우기)
np.std(X) # 표준편차
X.var()
X.std()
```

```python
# pandas 최빈값 구하기
X.mode()
X.astype(str).describe()["top"] # object 타입으로 바꿔 describe를 이용해 최빈값 구하기
```

```python
new_df = df.groupby('발행년도').count().번호.reset_index()
new_df.loc[new_df.발행년도 == '2014',:]

df.발행년도.value_counts().index.str.slice(0,4)
df.발행년도.value_counts().index.str[:4]

new_df = df.발행년도.value_counts() # 발행년도별로 count를 구하기 (Series)
new_df = new_df.reset_index() # 발행년도가 인덱스인 Series를 발행년도가 컬럼으로 된 데이터프레임으로 변환
new_df.columns =['발행년도','출판물수'] # 컬럼 이름 재정리
new_df.발행년도 = new_df.발행년도.str[:4] # 발행년도를 앞 4자리만 나오게 슬라이싱

# 앞 4자리가 int형이 아닌 애들을 지우기
remove_index = []
for idx, val in enumerate(new_df.발행년도):
    try:
        int(val)
    except:
        remove_index.append(idx)
new_df.drop(remove_index, axis=0, inplace=True)

new_df.발행년도 = new_df.발행년도.astype(int) # 발행년도를 int형으로 바꾸기

# 1900년도 이전, 2024년 이후 발행년도 데이터 제거
tidy_df = new_df.loc[(new_df.발행년도 > 1900) & (new_df.발행년도 < 2024),:]
```

```python
plt.figure(figsize=(10,10))
x = tidy_df.sort_values(by="발행년도").발행년도 # 발행년도를 오름차순으로 정렬
y = tidy_df.sort_values(by="발행년도").출판물수 # 출판물수를 발행년도 기준으로 오름차순 정렬
plt.plot(x,y)
plt.show()
```

```python
# 주제별 도서 개수 구하기
df.주제분류번호.fillna('Nan',inplace=True) # NaN 값을 'NaN' 으로 변환
```

```python
# 첫번째 값이 대분류라고 가정하고
new_df = df.주제분류번호.str.slice(0,1).value_counts().reset_index()
```

```python
new_df.columns = ['주제분류번호','개수'] # 컬럼 명 바꾸기
```

```python
tidy_df = new_df.loc[new_df.주제분류번호 != 'N',:] # 주제분류번호가 N인 값 제외해서 슬라이싱
```

```python
tidy_df.주제분류번호 = '주제분류번호-' + tidy_df.주제분류번호 # 주제분류번호 이름 바꾸기
```

```python
tidy_df.주제분류번호 = tidy_df.주제분류번호.str.replace('주제분류번호','subject') # 한글 깨지니까 영어로
```

```python
tidy_df.주제분류번호 = tidy_df.주제분류번호.str.capitalize() # 앞 글자만 대문자로 바꾸기
```

```python
x = tidy_df.loc[tidy_df.주제분류번호.sort_values().index,:].주제분류번호 # 주제분류번호로 정렬해서 인덱스 값 구하기
y = tidy_df.loc[tidy_df.주제분류번호.sort_values().index,:].개수
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,y,linestyle='-.',color='green',marker='.') # 종류 :  - / : / -- / -.
plt.show()
```

```python
# 발행년도 4자리 기준 각각 발행년도별 count
new_df = df.발행년도.str.slice(0,4).value_counts().reset_index()
```

```python
# 발행년도, 발행년도별 count로 구성된 데이터 프레임
new_df.columns = ['발행년도','발행부수']
```

```python
# abnormally 데이터 정리를 위한 인덱스 가져오기
remove_idx = []
for idx, val in enumerate(new_df.발행년도):
    try:
        int(val)
    except:
        remove_idx.append(idx)
```

```python
# remove 대상 인덱스를 제거
tidy_df = new_df.copy()
tidy_df.drop(remove_idx,0,inplace=True)
tidy_df.발행년도 = tidy_df.발행년도.astype(int)
```

```python
# 년도별 이상치 정리 (1945~2023)외의 데이터를 이상치
clean_df = tidy_df.loc[(tidy_df.발행년도 >= 1945) & (tidy_df.발행년도 <= 2023),:]
```

```python
# line 그래프로 그리기 위해 발행년도 순으로 정리
clean_df = clean_df.loc[clean_df.발행년도.sort_values().index,:]
```

```python
plt.figure(figsize=(10,7))
plt.plot(clean_df.발행년도,clean_df['발행부수'],color='green',linestyle=':',marker="X")
plt.title("Books by Year")
plt.xlabel("Year")
plt.ylabel("Number of Books")
plt.xticks(range(1947,2023,10)) #열의 시작 위치
for i in range(0,len(clean_df.index),5):
    plt.annotate(clean_df.iloc[i,1], (clean_df.iloc[i,0], clean_df.iloc[i,1]))
plt.show()
```
